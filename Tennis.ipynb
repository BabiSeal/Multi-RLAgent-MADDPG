````python
from maddpg import MADDPG
import torch
import numpy as np
import os
from collections import deque
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
print(f'\r Starting training')

def train_maddpg(env, brain_name, agent, solved_threshold, num_agents, n_episodes=5000, max_t=1000, print_every=10) :
    
    # track the average score (across all agents) for each episode
    scores_episode_history = []
    scores_deque = deque(maxlen=100)
    scores_avg = []
    
    # Iterate till max_episodes
    for i_episode in range(1, n_episodes+1):
        
        #Reset the Unity Environment
        env_info = env.reset(train_mode=True)[brain_name]
        #scores_debug = np.zeros(num_agents) 
        
        #Get the state information 
        states = env_info.vector_observations
        
        #reset agent
        agent.reset()
        
        #initialize current episode's all agent scores
        current_episode_scores = np.zeros(num_agents)
        rewards_array = []
        
        #Training loop for each epsisode
        for t in range(max_t):
            #determine the actions for each of the agents based on current states
            actions = agent.act(states)
            
            # actuate the actions on the environment to get next states, rewards to help build the experience info
            env_info = env.step(actions)[brain_name]
            next_states = env_info.vector_observations
            rewards = env_info.rewards
            rewards_array.append(rewards)
            current_episode_scores += rewards
            local_done = env_info.local_done
            
            #scores_debug += env_info.rewards
        
            
            # Now send the experience info to the state to save in replay buffer and sample from buffer to learn
            agent.step(states, actions, rewards, next_states, local_done)
            
            #Update state and current_episode scores
            states = next_states
        
            #print(f'\r t:{t} actions:{actions} rewards:{rewards_array} maxscores:{np.max(scores_debug)}')
            
            
            # If any agent is done, then the episode is done
            if np.any(local_done):
                #print(f'\r t:{t} local done rewards:{rewards_array} maxscores:{np.max(scores_debug)}')
                break 

        #update the episode's scores, calculate the mean over last 100 epsiode
        #episode_reward = np.max(current_episode_scores)  # Use max score across agents for this episode)
        episode_reward = np.max(np.sum(np.array(rewards_array),axis=0))
        
        scores_episode_history.append(episode_reward)
        scores_deque.append(episode_reward)
    
        if i_episode % print_every == 0:
            print(f'\rEpisode {i_episode}\tAverage Score: {np.mean(scores_deque):.4f}')
            print("\n")
        if np.mean(scores_deque) >= solved_threshold:
            print(f'\nEnvironment solved in {i_episode-100} episodes!\tAverage Score: {np.mean(scores_deque):.4f}')
            for i, a in enumerate(agent.agents):
                torch.save(a.actor_local.state_dict(), f'actor_agent_{i}.pth')
                torch.save(a.critic_local.state_dict(), f'critic_agent_{i}.pth')
            break
        
    return scores_episode_history
````

````python
from maddpg import MADDPG
import os
os.environ["CUDA_VISIBLE_DEVICES"] = ""     # hide all GPUs from PyTorch

number_of_episodes = 6000
max_t = 1000
print_every = 50

batch_size = 256

noise_weight = 0.5
noise_reduction = 1.0
solved_threshold = 0.5

UPDATE_EVERY= 2

# reset the environment
env_info = env.reset(train_mode=True)[brain_name]

# number of agents 
num_agents = len(env_info.agents)

# size of each action
action_size = brain.vector_action_space_size
print('Size of each action:', action_size)

# examine the state space 
states = env_info.vector_observations
state_size = states.shape[1]

agent = MADDPG(state_size=state_size, action_size=action_size, num_agents=num_agents, seed=0, noise_weight=0.5, noise_decay=noise_reduction, update_every=UPDATE_EVERY)
scores = train_maddpg(env, brain_name, agent, solved_threshold=0.5, num_agents=2, n_episodes=number_of_episodes, max_t=max_t, print_every=print_every)

````

````python
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(1, len(scores)+1), scores)
plt.ylabel('Score')
plt.xlabel('Episode #')
plt.show()
````
